# Favorite Project: MediPulse
My favorite project this year has to be MediPulse.
This project felt like the perfect combination of challenge and purpose. We were building something beyond for grades, it was about creating something that could actually help people make smarter healthcare decisions. My role in my team focused on frontend designs including artifacts and backend features with API. 

What made it even more exciting was that this project connected me to a real world opportunity through PilotCity giving us a chance of internship. Although, I won't be able to continuing to develop MediPulse beyond class, adding more features and pushing it even further.
