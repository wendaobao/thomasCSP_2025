```python
%%html
<html>
<head>
    <title>Treasure Hunt Game</title>
    <style>
        table {
            border-collapse: collapse;
        }
        td {
            width: 30px;
            height: 30px;
            text-align: center;
            border: 1px solid #000;
            font-family: monospace;
        }
        .button-container {
            margin-top: 10px;
        }
    </style>
</head>
<body>
    <h1>Treasure Hunt Game</h1>
    <table id="game-map"></table>
    <div class="button-container">
        <button onclick="movePlayer('up')">Up</button>
        <br>
        <button onclick="movePlayer('left')">Left</button>
        <button onclick="movePlayer('right')">Right</button>
        <br>
        <button onclick="movePlayer('down')">Down</button>
    </div>
    <div id="npc-dialog"></div>
    
    <script>
        let gameMap = [
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "], // Treasure hidden
            [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "]
        ];

        let playerPosition = [0, 0];
        let npcPosition = [Math.floor(Math.random() * 10), Math.floor(Math.random() * 10)];
        let treasurePosition = [8, 3]; // Treasure's position

        const npcDialog = "I've heard rumors of a treasure nearby!";
        const npcHint = "It's located at row 9, column 4.";

        // Display the game map as a table
        function displayGameMap() {
            let table = document.createElement("table");
            for (let i = 0; i < gameMap.length; i++) {
                let row = document.createElement("tr");
                for (let j = 0; j < gameMap[i].length; j++) {
                    let cell = document.createElement("td");
                    if (playerPosition[0] === i && playerPosition[1] === j) {
                        cell.textContent = "P"; // P for Player
                    } else if (npcPosition[0] === i && npcPosition[1] === j) {
                        cell.textContent = "N"; // N for NPC
                    } else if (i === treasurePosition[0] && j === treasurePosition[1]) {
                        cell.textContent = " "; // Keep treasure hidden
                    } else {
                        cell.textContent = gameMap[i][j];
                    }
                    row.appendChild(cell);
                }
                table.appendChild(row);
            }
            let gameMapDiv = document.getElementById("game-map");
            gameMapDiv.innerHTML = "";
            gameMapDiv.appendChild(table);
        }

        // Handle player movement
        function movePlayer(direction) {
            if (direction === "up" && playerPosition[0] > 0) {
                playerPosition[0] -= 1;
            } else if (direction === "down" && playerPosition[0] < 9) {
                playerPosition[0] += 1;
            } else if (direction === "left" && playerPosition[1] > 0) {
                playerPosition[1] -= 1;
            } else if (direction === "right" && playerPosition[1] < 9) {
                playerPosition[1] += 1;
            }

            // Check if player encounters NPC
            if (playerPosition[0] === npcPosition[0] && playerPosition[1] === npcPosition[1]) {
                interactWithNpc();
            }

            // Check if player finds the treasure (after NPC hint)
            if (playerPosition[0] === treasurePosition[0] && playerPosition[1] === treasurePosition[1]) {
                alert("Congratulations! You've found the treasure!");
            }

            displayGameMap(); // Update the game map display
        }

        // Interaction with NPC
        function interactWithNpc() {
            let dialogDiv = document.getElementById("npc-dialog");
            dialogDiv.textContent = npcDialog;

            setTimeout(() => {
                alert(npcHint); // Show NPC hint after dialog
                revealTreasure(); // Reveal the treasure
            }, 500);
        }

        // Reveal the treasure on the map
        function revealTreasure() {
            gameMap[treasurePosition[0]][treasurePosition[1]] = "T"; // Show treasure
            displayGameMap(); // Update the game map display
        }

        // Display the initial game map
        displayGameMap();
    </script>
</body>
</html>
```

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Sprite Sheet Animation</title>
    <style>
        body {
            display: flex;
            flex-direction: column;
            align-items: center;
        }
        canvas {
            border: 1px solid #000;
        }
    </style>
</head>
<body>
    <h2>Sprite Sheet Animation</h2>
    <canvas id="spriteCanvas"></canvas>
    
    <script>
        let img = new Image();
        let currentFrame = 0;
        let frameCount = 0;
        let ctx, canvas;

        // URL of the sprite sheet
        const spriteUrl = 'https://static.wixstatic.com/media/c6fb96_88ba297597bc4ad0812e06be20cb28c9~mv2.png/v1/fill/w_1600,h_900,al_c,q_90/file.jpg';
        const frameWidth = 64;      // Width of each sprite frame
        const frameHeight = 64;     // Height of each sprite frame
        const totalFrames = 16;     // Total number of frames in the sprite sheet
        const frameSpeed = 10;       // Speed of the animation (frames to wait before changing)

        function startAnimation() {
            canvas = document.getElementById("spriteCanvas");
            canvas.width = frameWidth;
            canvas.height = frameHeight;
            ctx = canvas.getContext("2d");
            img.src = spriteUrl;
            img.onload = displaySprite;
        }

        function displaySprite() {
            const col = currentFrame % (totalFrames / 4); // Calculate column in the sprite sheet
            const row = Math.floor(currentFrame / (totalFrames / 4)); // Calculate row in the sprite sheet
            
            ctx.clearRect(0, 0, canvas.width, canvas.height); // Clear the canvas
            ctx.drawImage(img, col * frameWidth, row * frameHeight, frameWidth, frameHeight, 0, 0, frameWidth, frameHeight); // Draw the current frame

            frameCount++;
            if (frameCount >= frameSpeed) {
                currentFrame = (currentFrame + 1) % totalFrames; // Move to the next frame
                frameCount = 0; // Reset frame count
            }
            requestAnimationFrame(displaySprite); // Request next frame
        }

        window.onload = startAnimation; // Start animation on window load
    </script>
</body>
</html>
```
