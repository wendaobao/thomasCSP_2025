### Big Idea 4: Computer Systems and Networks (CSN)

- Explores how computers and networks connect, communicate, and process information.
- Key ideas include how the Internet works, routing, and how systems handle failure.

**Exam Weight**: 11%–15%

#### 4.1 The Internet

- Devices like computers and routers form networks that exchange data.
- Data is sent across paths using a process called routing.
- Internet protocols are open standards that guide communication.
- The Internet was designed to scale with demand.

#### 4.2 Fault Tolerance

- A fault-tolerant system keeps working even when parts fail.
- Redundancy (extra hardware or pathways) increases reliability.
- Network redundancy allows data to reroute when paths fail.
- This supports system growth and minimizes disruptions.

#### 4.3 Computing Models & Efficiency

- **Sequential Computing**: Tasks run one by one in a specific order.
- **Parallel Computing**: Breaks tasks into parts that run at the same time. More efficient but limited by sequential tasks (Amdahl’s Law).
- **Distributed Computing**: Uses multiple machines to solve bigger problems faster.
- Speedup = how much faster a parallel solution is compared to doing everything sequentially.

- Includes issues like access to technology (digital divide), algorithmic bias, and online safety.
